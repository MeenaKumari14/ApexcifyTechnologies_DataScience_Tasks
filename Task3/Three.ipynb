```python
#  TASK 3 â€” PREMIUM EDITION
#  Dataset Exploration EDA

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 13

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/First_Internship/Task3/iris_like.csv")

print("DATASET LOADED SUCCESSFULLY")
print(df.head(), "\n")
```

```python
# 1 Dataset Overview

print("DESCRIPTIVE STATISTICS")
print(df.describe(), "\n")

print("MISSING VALUES")
print(df.isnull().sum(), "\n")

print("LABEL DISTRIBUTION")
print(df["Label"].value_counts(), "\n")

```

```python
# 2 Class Distribution (Bar Chart)

plt.figure(figsize=(8, 5))
sns.countplot(x="Label", data=df, palette="Blues")
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
# 3 Correlation Heatmap

plt.figure(figsize=(8, 6))
sns.heatmap(
    df.drop("Label", axis=1).corr(),
    annot=True,
    cmap="Blues",
    fmt=".2f"
)
plt.title("Feature Correlation Heatmap")
plt.show()

```

```python
# 4 Pairplot (Signature Data Science Plot)

sns.pairplot(
    df,
    hue="Label",
    palette="deep",
    diag_kind="kde"
)
plt.show()
```

```python
# 5) Feature Distributions by Class
features = ["Length1", "Length2", "Width1", "Width2"]

for feature in features:
    plt.figure(figsize=(8, 5))
    sns.kdeplot(
        data=df,
        x=feature,
        hue="Label",
        fill=True
    )
    plt.title(f"Distribution of {feature} by Class")
    plt.show()
```

```python
# 6) Boxplots (Outlier & Spread Analysis)
for feature in features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(
        x="Label",
        y=feature,
        data=df,
        palette="Blues"
    )
    plt.title(f"{feature} Comparison Across Classes")
    plt.show()

```

```python
# 7) Scatter Relationship (Feature Interaction)

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="Length1",
    y="Width1",
    hue="Label",
    palette="deep",
    s=80
)
plt.title("Length1 vs Width1 Relationship")
plt.show()

```
